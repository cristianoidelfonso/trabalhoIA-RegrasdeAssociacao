## Mostre o lift de todas as Regras de Associação (utilizando o algoritmo de Apriori).

## Utilize suporte mínimo de 0,3 e confiança mínima 0,9

```python
# importação de bibliotecas e configuração do ambiente
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#lendo o arquivo csv de coletas de dados da pesquisa
dadosAtividadeFisica = pd.read_csv('praticaAtvFisica.csv', encoding="utf8") 
#calculando o suporte
suportesAtividadeFisica = apriori(dadosAtividadeFisica, min_support=0.3, use_colnames=True)
#imprime os suportes
suportesAtividadeFisica 
```

```python
#mandando calcular as regras com confianca minima 0.9
regrasAtividadeFisica = association_rules(suportesAtividadeFisica, metric="confidence", min_threshold=0.9)
regrasAtividadeFisica
```

## Perfil: Homens, entre 20 e 30 anos, praticam atividades físicas em academia, acima de tres vezes por semana, no período noturno e ja sofreram alguma lesão durante o treino.
